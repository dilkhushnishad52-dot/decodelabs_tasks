### TASK 2: Data Cleaning & Preprocessing



```python
import pandas as pd
import numpy as np

df = pd.read_excel('Online-Store-Orders.xlsx')

print(df.isnull().sum())    #309 missing values in couponcode

# fill missing data 
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

#df.duplicated().sum()        # no  duplicates data

# remove duplicates
df.drop_duplicates(inplace=True)

#print(df['Date'].dtypes)     # already in datatime formate
df['Date'] = pd.to_datetime(df['Date'])

# cleaned dataset
print(df.shape)

```
